```python
import matplotlib.pyplot as plt

import pandas as pd

df = pd.read_excel (r'./Esports.xlsx')
top10 = df.nlargest(10,'TotalUSDPrize')
top10.plot(x="TeamName", y="TotalUSDPrize")
```

```python
top10 = df.nlargest(10,'TotalTournaments')
top10.plot.bar(x="TeamName", y="TotalTournaments")
```

```python
genreDF = df.groupby('Genre').count()
```
